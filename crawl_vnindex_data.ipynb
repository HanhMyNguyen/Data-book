```python
import requests
import pandas as pd
import pyodbc
```

```python
response = requests.get("https://s.cafef.vn/Ajax/PageNew/DataHistory/PriceHistory.ashx?Symbol=VNINDEX&StartDate=01/01/2020&EndDate=02/29/2020&PageIndex=1&PageSize=20")
```

```python
data_vnindex_json = response.json()
```

```python
data_vnindex_json['Data']['Data']
```

```python
data_vnindex = []
for i in data_vnindex_json['Data']['Data']:
  data_vnindex.append([i['Ngay'], i['GiaDieuChinh'], i['GiaDongCua'], i['ThayDoi'],i['KhoiLuongKhopLenh'], i['GiaTriKhopLenh'], i['KLThoaThuan'], i['GtThoaThuan'], i['GiaMoCua'], i['GiaCaoNhat'], i['GiaThapNhat']])
```

```python
data_vnindex_df = pd.DataFrame(data_vnindex, columns = ['Ngay','GiaDieuChinh','GiaDongCua','ThayDoi','KhoiLuongKhopLenh','GiaTriKhopLenh','KLThoaThuan', 'GtThoaThuan','GiaMoCua','GiaCaoNhat','GiaThapNhat' ])
```

```python
file_path = r'D:\Hanhntm36\Handover\_MyFolder\7. Analysis\2024\CASA key drivers\VNINDEX.csv'
# file_path = 'VNINDEX.csv'
data_vnindex_df.to_csv(file_path, index=False)
```
